```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

df = pd.read_csv("movies.csv")

plt.figure(figsize=(10, 6))
sns.regplot(data=df, x="numVotes", y="averageRating",
            scatter=False)  # geen punten meer!

plt.xscale("log")
plt.title("Aantal stemmen vs gemiddelde beoordeling (zonder punten)")
plt.xlabel("Aantal stemmen (log-schaal)")
plt.ylabel("Gemiddelde beoordeling")
plt.tight_layout()
plt.show()
```
